```python
#@title Genre Information

options = ['Hip-Hop', 'Pop', 'Rock', 'Folk', 'Jazz', 'Electronic', 'Spoken', 'International',
  'Soul-RnB', 'Blues', 'Country', 'Classical', 'Old-Time / Historic', 'Instrumental', 'Easy Listening']

# Initialize an empty dictionary to store track_ids for each genre
genre_track_ids = {}

for genre in options:
    # Filter track_ids for the current genre
    genre_track_ids[genre] = list(tracks_mod.loc[tracks_mod['genre_top'] == genre, 'track_id'])

# print("Track IDs for Rock genre:", genre_track_ids['Rock'])


# # #experiment to see if genre selection works
# selected_genre = "Rock"
# selected_genre_track_ids = genre_track_ids[selected_genre]
# selected_results = results[results['track_id'].isin(selected_genre_track_ids)]

# selected_results
```

```python
#@title Filter by Genre
print("Use the Dropdown Menu to filter by Genre\nKeep in mind that not all genres are recommended with a given song")

# Create a dropdown widget
dropdown = widgets.Dropdown(options=options, description='')

# Variable to store the selected value
selected_value = "Classical"

# Define a function to be called when the dropdown value changes
def on_dropdown_change(change):
    global selected_value
    selected_value = change['new']
    print(f'Selected option: {selected_value}')

    # Update the dataframe based on the new selected genre
    selected_genre_track_ids = genre_track_ids[selected_value]
    selected_results = results[results['track_id'].isin(selected_genre_track_ids)]

    # Clear previous output and display the updated dataframe
    with output_area:
        clear_output(wait=True)
        display(selected_results)

# Attach the function to the dropdown's value change event
dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown widget
display(dropdown)

# Create an output area for the dataframe
output_area = widgets.Output()
display(output_area)

# Initial filtering and display
selected_genre_track_ids = genre_track_ids[selected_value]
selected_results = results[results['track_id'].isin(selected_genre_track_ids)]
with output_area:
    display(selected_results)
```
