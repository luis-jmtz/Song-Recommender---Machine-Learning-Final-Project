```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import joblib  # Import joblib for loading the model
from google.colab import drive
```

```python
# Load the trained KMeans model
kmeans = joblib.load("/content/drive/Shareddrives/Machine Learning Group Project/Final Google Colabs/final_k_cluster_model_with_250_clusters")
```

```python
new_data = pd.read_csv("/content/drive/Shareddrives/Machine Learning Group Project/Final Google Colabs/Copy of stressed_out_echonest_features.csv")
if 'Liveness' in new_data.columns:
  new_data = new_data.drop('Liveness', axis=1)
```

```python
new_data.head()
```

```python
# Extract features for prediction, ignoring the first column
new_features = new_data.iloc[:, 1:].values
```

```python
# Create a new scaler and fit it to the new data
scaler = StandardScaler()
scaler.fit(new_features)
```

```python
# Standardize the features for prediction
new_features_scaled = scaler.transform(new_features)
```

```python
new_data['Cluster'] = kmeans.predict(new_features_scaled)
```
