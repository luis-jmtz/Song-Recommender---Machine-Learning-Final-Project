```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from google.colab import drive
import joblib
```

```python
# # Mount Google Drive to access files
# drive.mount('/content/drive')
```

```python
# Load the CSV file into a Pandas DataFrame
echonest = pd.read_csv("/content/drive/Shareddrives/Machine Learning Group Project/echonest_mod4.csv")
```

```python
echonest.head()
```

```python
if 'Unnamed: 0' in echonest.columns:
  echonest = echonest.drop('Unnamed: 0', axis=1)

if 'liveness' in echonest.columns:
  echonest = echonest.drop('liveness', axis=1)
```

```python
echonest.head()
```

```python
echonest.shape
```

```python
features = echonest.iloc[:, 1:].values  # we are not taking into account the ID
```

```python
# Standardize the features for better performance of the KMeans algorithm
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# train the KMeans model
num_clusters = 250
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(features_scaled)
```

```python
# Save the KMeans model to a file using joblib
# joblib.dump(kmeans, "/content/drive/Shareddrives/Machine Learning Group Project/Final Google Colabs/final_k_cluster_model_with_250_clusters")
```

```python
echonest_clusters_version_1_25_clusters = echonest
```

```python
echonest_clusters_version_1_25_clusters['Cluster'] = kmeans.labels_
```
