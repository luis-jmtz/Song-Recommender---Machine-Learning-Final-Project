```python
import pandas as pd
import joblib
```

```python
# Load the pre-trained k-means model
loaded_model = joblib.load("/content/drive/Shareddrives/Machine Learning Group Project/features_kmeans_25_cluster_model.joblib")
```

```python
new_data = pd.read_csv("/content/drive/Shareddrives/Machine Learning Group Project/stressedout_features.csv")
```

```python
# Predict the cluster for the new data
new_data['cluster'] = loaded_model.predict(new_data)
```

```python
# Display the assigned cluster for the new data
assigned_cluster = new_data['cluster'].iloc[0]
print(f"Assigned Cluster for New Data: {assigned_cluster}")
```
