```python
def frameCalc(specific_string):
    # List to store results
    result_data = []

    # Iterate through each row in the dataframe
    for row_index in range(len(features)):
        # Find columns containing the specific string
        selected_columns = [col for col in features.columns if specific_string in col]

        # Filter the DataFrame to include only the selected columns and the specific row
        selected_data = features.loc[row_index, selected_columns]

        # Calculate mean, median, and range for the selected row
        mean_value = selected_data.mean()
        median_value = selected_data.median()
        range_value = selected_data.max() - selected_data.min()
        min_value = selected_data.min()
        max_value = selected_data.max()

        # Append the results to the list
        result_data.append([mean_value, median_value, range_value, min_value, max_value])

        # Append the results as new columns to the existing DataFrame
        features_2.loc[row_index, specific_string + " Mean"] = mean_value
        features_2.loc[row_index, specific_string + " Median"] = median_value
        features_2.loc[row_index, specific_string + " Range"] = range_value
        features_2.loc[row_index, specific_string + " Min"] = min_value
        features_2.loc[row_index, specific_string + " Max"] = max_value
```
