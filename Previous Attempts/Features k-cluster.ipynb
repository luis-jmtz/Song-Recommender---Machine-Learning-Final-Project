```python
import numpy as np
import pandas as pd
import librosa
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from joblib import dump
```

```python
# Load data
features = pd.read_csv("/content/drive/Shareddrives/Machine Learning Group Project/encoded_features.csv")
```

```python
features.head()
```

```python
# features = features.drop('cluster', axis=1)
```

```python
X = features.values  # Assuming all columns are features

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
optimal_k = 100
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X_scaled)
```

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce dimensionality to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add the cluster assignments to the reduced data
X_with_clusters = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
X_with_clusters['cluster'] = kmeans.labels_

# Visualize the clusters
plt.figure(figsize=(10, 6))
for cluster in range(optimal_k):
    cluster_points = X_with_clusters[X_with_clusters['cluster'] == cluster]
    plt.scatter(cluster_points['PC1'], cluster_points['PC2'], label=f'Cluster {cluster + 1}')

plt.title('KMeans Clustering Result (2D PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
model_path = "/content/drive/Shareddrives/Machine Learning Group Project/features_kmeans_model.joblib"
dump(kmeans, model_path)
```
